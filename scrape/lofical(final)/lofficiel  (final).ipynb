```python
Title = []
Date = []
URLs = []
Category = []
URLs = []
Link = []
Details=[]
```

```python
for i in range(1,21):
    web = 'https://www.lofficielusa.com/search?query=jewelry' + str(i)
    driver.get(web)
    b = soup(driver.page_source,'html.parser')
    news = b.find_all('div',{'class':'space--bottom-1-on-lap'})
    
    for new in news:
        try:
            Title.append(new.find('span',{'class':'hover-underline'}).get_text())
        except:
            Title.append('n/a')
        try:
            Date.append(new.find('div',{'class':'stack__footer'}).get_text())
        except:
            Date.append('n/a')
        try:
            URLs.append(new.find('a',{'class':'stack'}).get('href'))
        except:
            URLs.append('n/a')
```

```python
url_fashion = 'https://www.lofficielusa.com'
for j in URLs:
    print(url_fashion + j)
    Links = requests.get(url_fashion + j)
    b2 = soup(Links.text,'html.parser')
    
    for text_body in b2.find_all('main',{'class':'site-content'}):
        top = text_body.text.replace('\xa0','')
        if top != None:
            Details.append(top)
```

```python
table = pd.DataFrame([Title, Date ,Details]).transpose()
table.columns = ['Title','Date','Details']
table
```
