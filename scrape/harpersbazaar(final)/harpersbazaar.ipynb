```python
Title = []
Date = []
URLs = []
Link = []
Details=[]
```

```python
while True:
    try:
        loadmore = driver.find_element_by_css_selector("body > main > div.feed.feed-grid > button").click()
    except NoSuchElementException:
        break
    b = soup(driver.page_source,'html.parser')
    news = b.find_all('div',{'class':'simple-item grid-simple-item'})  
    
    for new in news:
        try:
            Title.append(new.find('div',{'class' : 'simple-item-title item-title'}).get_text())
        except:
            Title.append('n/a')
        try:
            URLs.append(new.find('a').get('href'))
        except:
            URLs.append('n/a')
```

```python
url_fashion = 'https://www.harpersbazaar.com'
for j in URLs:
    print(url_fashion + j)
    Links = requests.get(url_fashion + j)
    b2 = soup(Links.text,'html.parser')
    
    #temp = []
    for text_body in b2.find_all('div',{'class':'article-body-content article-body standard-body-content css-z6i669 ewisyje5'}):
        top = text_body.text.replace('\xa0','')
        if top != None:
            Details.append(top)
```

```python
print(Details)
```

```python
table = pd.DataFrame([Title,Details]).transpose()
table.columns = ['Title','Details']
table
```

```python
df = pd.DataFrame(table)
df
```
